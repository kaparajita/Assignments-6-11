```python
ds = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data',header = None)
ds.columns = ['age','workclass','fnlwgt','education','education-num','marital_status','occupation',
              'relationship','race','sex','capital_gain','capital_loss','hours_per_week','native_country','state']
ds
```

```python
#1. Create an sqlalchemy engine using a sample from the data set
import sqlalchemy as sa
engine = sa.create_engine('sqlite:///mydb1', echo =False)
ds.to_sql('ds', engine, if_exists='replace', chunksize=1000)
```

```python
#2. Write two basic update queries
# Print to see the existing values
result = engine.execute("SELECT fnlwgt, education,age FROM ds WHERE age = 39").fetchall()
print("{l} records found".format(l=len(result)))
print(result)
print("\n")
```

```python
#update adult table recordes of ecduation values to Technical where age is 39
engine.execute("UPDATE ds set education = 'Technical' WHERE age = 39")

#update adult table of fnlwgt columns values to 0 where age  = 39
engine.execute("UPDATE ds set fnlwgt = 0 WHERE age = 39")
```

```python
#Checking Reslut below: Updated education and fnlwgt column values.
result = engine.execute("SELECT fnlwgt, education,age FROM ds WHERE age = 39").fetchall()
print("{l} records found".format(l=len(result)))
print(result)
print("\n")
```

```python
#2.Two Delete Queries
#Delete queries to delete records where age is 39 and fnlwgt is 128392
engine.execute("DELETE FROM ds WHERE age = 39")
engine.execute("DELETE FROM ds WHERE fnlwgt = 128392")
```

```python
#Check the table for deleted records
result = engine.execute("SELECT fnlwgt, education,age FROM ds WHERE age = 39").fetchall()
print("{l} records found".format(l=len(result)))
print(result)
print("\n")
```

```python
#3.Two Filter Queries

result = engine.execute("SELECT fnlwgt, education,age FROM ds WHERE age = {age}".format(age=50)).fetchall()
result1 = engine.execute("SELECT fnlwgt, education FROM ds WHERE education = '{edu}'".format(edu="Bachelors")).fetchall()
```

```python
#5.Write Two Function Queries
def fib(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b

# Create a new table with Column named "Value"
engine.execute('CREATE TABLE IF NOT EXISTS Fib (value INTEGER)')
engine.execute('INSERT INTO Fib VALUES (?)',[(str(x),) for x in fib(10)])
print(engine.execute("SELECT * FROM Fib").fetchall())
```

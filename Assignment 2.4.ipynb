```python
#2.4. Problem Statement: Matplotlib
#Matplotlib:
#This assignment is for visualization using matplotlib:
#data to use:
#url=https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv

import pandas as pd
titanic = pd.read_csv("https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv")
titanic
```

```python
#1. Create a pie chart presenting the male/female proportion

import matplotlib.pyplot as plt
%matplotlib inline

femalecount = len(titanic[titanic['sex'] == 'female'])
malecount  = len(titanic[titanic['sex'] == 'male'])
genderlist = [malecount,femalecount]
colours = ['blue','red']
labels = ['male','female']
plt.pie(genderlist,colors = colours,labels=labels,autopct='%1.0f%%',)
plt.show()
```

```python
#2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

fares = titanic['fare'].values
age = titanic['age'].values
colours = titanic['sex']
plt.scatter(fares,age ,norm = 'Normalize')
plt.show()
```
